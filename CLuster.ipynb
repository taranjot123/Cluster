```python
import pandas as pd
```

```python
data=pd.read_csv('xclara.csv')
```

```python
data.head()
```

```python
import numpy as np
```

```python
c1=data['V1'].values
```

```python
c2=data['V2'].values
```

```python
from sklearn.cluster import KMeans
```

```python
range_cluster = [1,2,3,5,7,10,15]
```

```python
sse={}
# dictionary
```

```python
for n_clusters in range_cluster:
    kmeans=KMeans(n_clusters=n_clusters)
    kmeans_label=kmeans.fit(data)
    sse[n_clusters]=kmeans_label.inertia_
    #inertia is the score function for Kmeans
    #inertia finds the fitting 
```

```python
from matplotlib import pyplot as plt
#plt.figure()
plt.plot(list(sse.keys()),list(sse.values()))
plt.xlabel('Number of Cluster')
plt.ylabel('cost')
plt.show()
#Prefered K is seen to be as 3
```

```python
plt.scatter(c1,c2,c='yellow') #represents the enrties as a cluster 
```

```python
kmeans=KMeans(n_clusters=3)# any numnber can be putt here <the number will make the amount of clusters>
kmeans_label=kmeans.fit_predict(data)#helps to make the graph
```

```python
plt.scatter(c1,c2,c=kmeans_label)# are the 3 clusters which are to be plotted
```

```python
centroids=kmeans.cluster_centers_ #center points of the clusters in the graph
```
